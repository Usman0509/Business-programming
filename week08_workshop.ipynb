## 📁 Loading Sample Data

First, let's load our sample transaction data:

````python
import pandas as pd

# Load the sample transactions
df = pd.read_csv('data/transactions.csv')

# Display the first few rows to understand our data
print("📋 Sample Transaction Data:")
print("=" * 40)
print(df.head())

print(f"\n📊 Dataset Info:")
print(f"Total transactions: {len(df)}")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Categories: {df['Category'].unique()}")
````

## 🤖 AI Collaboration Exercise 1: Basic Analysis

**Your Turn**: Practice prompting AI for basic data analysis.

::: {.callout-tip}
## Practice Prompt Template

Try asking AI:
*"Using this transaction DataFrame, I want to understand my spending patterns. Please create code that shows:
1. Total spending by category
2. Average transaction amount by category  
3. Number of transactions per category
Format the output to be easy to read for a business presentation."*
:::

**Expected AI Response** (try this yourself first!):

````python
# AI-generated solution for spending analysis

# Calculate comprehensive spending summary using named aggregation
# This method correctly renames the aggregated columns and leverages the pre-cleaned 'Amount_clean' column.
spending_analysis = df.groupby('Category').agg(
    Total_Spent=('Amount_clean', 'sum'),
    Avg_Amount=('Amount_clean', 'mean'),
    Transaction_Count=('Amount_clean', 'count')
).round(2)

# Sort by total spending
symmetric_diff_analysis = spending_analysis.sort_values('Total_Spent', ascending=False)

print("💼 Business Spending Analysis")
print("=" * 50)
print(f"{'Category':<15} {'Total':<10} {'Average':<10} {'Count':<8}")
print("-" * 50)

for category, row in spending_analysis.iterrows():
    total = row['Total_Spent']
    avg = row['Avg_Amount']
    count = row['Transaction_Count']
    print(f"{category:<15} ${total:<9.2f} ${avg:<9.2f} {count:<8.0f}")

print(f"\nGrand Total: ${spending_analysis['Total_Spent'].sum():.2f}")
````

## 🤖 AI Collaboration Exercise 2: Business Insights

**Challenge**: Ask AI to help you identify spending patterns that would be useful for financial planning.

**Sample prompt to try:**
*"Based on this spending data, help me identify potential areas where I could reduce expenses. Look for categories with high spending or frequent transactions that might represent opportunities for savings."*

````python
# Your AI collaboration code goes here
# Try the prompt above and see what AI suggests!
# Clean Amount column once
df['Amount_clean'] = pd.to_numeric(df['Amount'].str.replace('$', '', regex=False), errors='coerce')

# Remove invalid rows
df = df.dropna(subset=['Amount_clean'])

# Spending totals by category
category_totals = df.groupby('Category')['Amount_clean'].sum().sort_values(ascending=False)

# Transaction counts by category
category_counts = df.groupby('Category')['Amount_clean'].count().sort_values(ascending=False)

# Average transaction by category
category_averages = df.groupby('Category')['Amount_clean'].mean().sort_values(ascending=False)

print("💡 Spending Insights for Financial Planning")
print("=" * 50)

print("\nTop spending categories:")
for category, total in category_totals.items():
    print(f"- {category}: ${total:.2f}")

print("\nMost frequent spending categories:")
for category, count in category_counts.items():
    print(f"- {category}: {count} transactions")

print("\nHighest average transaction categories:")
for category, avg in category_averages.items():
    print(f"- {category}: ${avg:.2f}")

print("\n📉 Potential Savings Opportunities:")
highest_category = category_totals.index[0]
highest_total = category_totals.iloc[0]
print(f"- The highest spending category is {highest_category} at ${highest_total:.2f}. This may be a good area to review for savings.")

most_frequent_category = category_counts.index[0]
most_frequent_count = category_counts.iloc[0]
print(f"- The most frequent transactions are in {most_frequent_category} ({most_frequent_count} transactions). Frequent small purchases may add up over time.")

highest_average_category = category_averages.index[0]
highest_average_amount = category_averages.iloc[0]
print(f"- The highest average transaction amount is in {highest_average_category} at ${highest_average_amount:.2f}, which may indicate larger one-off expenses to monitor.")
````

## 📊 Step 1: Set Up Your Data

**Open your assignment notebook** (`smart_finance_assistant.ipynb`) and start with this section:

### Data Loading and Exploration

::: {.callout-tip}
## Prompt AI for this section:

*"I'm building a Smart Finance Assistant for my programming assignment. I need to load a CSV file with transaction data (columns: Date, Amount, Category, Description). The Amount column has dollar signs that need cleaning. Please write professional Python code with detailed comments that:
1. Loads the CSV data
2. Cleans and validates the data
3. Shows basic information about the dataset
4. Handles any common data issues"*

````python
# AI-assisted data loading section
# (Copy the code AI provides here)
import pandas as pd

# Load transaction data
df = pd.read_csv("data/transactions.csv")

# Preview the data
print("📋 Transaction Data Preview")
print("=" * 40)
print(df.head())

# Clean Amount column by removing dollar signs
df["Amount"] = df["Amount"].str.replace("$", "", regex=False)

# Convert Amount to numeric
df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Remove invalid rows
df = df.dropna(subset=["Amount", "Date"])

# Show basic information
print("\n📊 Dataset Info")
print("=" * 40)
print(df.info())

print("\nUnique categories:")
print(df["Category"].unique())

print("\nSummary statistics:")
print(df.describe())

# Learning note:
# I learned how to load CSV data, clean dollar values, convert columns to correct types,
# and remove invalid rows before analysis.
# Add your own comments about what you learned
````

## 📈 Step 2: Create Core Analysis Functions

**Prompt AI** to help you create the foundation of your finance assistant:

::: {.callout-tip}
## Analysis Functions Prompt:

*"For my Smart Finance Assistant, I need functions to analyze spending data. Please create well-documented functions that:
1. Calculate total spending by category
2. Find the top 5 spending categories
3. Calculate average spending per transaction
4. Identify any unusual transactions (very high or negative amounts)
Each function should have clear business explanations and error handling."*
:::

````python
# AI-generated analysis functions
# (Your AI collaboration code goes here)
def total_spending_by_category(df):
    """Calculate total spending by category."""
    return df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

def top_5_spending_categories(df):
    """Return the top 5 spending categories."""
    return df.groupby("Category")["Amount"].sum().sort_values(ascending=False).head(5)

def average_spending_per_transaction(df):
    """Calculate the average transaction amount."""
    return df["Amount"].mean()

def unusual_transactions(df, high_threshold=500):
    """Identify unusually high or negative transactions."""
    return df[(df["Amount"] > high_threshold) | (df["Amount"] < 0)]

print("💼 Spending by Category")
print(total_spending_by_category(df))

print("\n🏆 Top 5 Categories")
print(top_5_spending_categories(df))

print(f"\n📊 Average Transaction Amount: ${average_spending_per_transaction(df):.2f}")

print("\n⚠️ Unusual Transactions")
print(unusual_transactions(df))
````

## 💰 Step 3: Generate Business Insights

**Challenge AI** to help you create business-relevant insights:

::: {.callout-tip}
## Business Insights Prompt:

*"Using the transaction analysis functions, help me create a comprehensive spending report that would be useful for personal financial planning. Include recommendations for potential savings and spending pattern observations. Format it like a professional financial summary."*
:::

````python
# AI-generated business insights
# (Your AI collaboration code goes here)
category_totals = total_spending_by_category(df)
avg_transaction = average_spending_per_transaction(df)
unusual = unusual_transactions(df)

print("📈 Personal Finance Summary")
print("=" * 50)

print(f"Total spending overall: ${df['Amount'].sum():.2f}")
print(f"Average transaction amount: ${avg_transaction:.2f}")

print("\nTop spending categories:")
for category, total in category_totals.head(5).items():
    print(f"- {category}: ${total:.2f}")

print("\nRecommendations:")
highest_category = category_totals.index[0]
highest_total = category_totals.iloc[0]
print(f"- Your biggest spending area is {highest_category} at ${highest_total:.2f}. This is a good category to review for possible savings.")

if len(unusual) > 0:
    print(f"- There are {len(unusual)} unusual transactions. These should be checked for one-off or unexpected expenses.")
else:
    print("- No unusual transactions were detected based on the current threshold.")

print("- Tracking category totals regularly can help improve budgeting and financial planning.")
````
