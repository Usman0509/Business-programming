### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]
income_str = input("Enter your monthly income: ")
income = float(income_str.replace('$', ''))

expense1_str = input("Enter expense 1: ")
expense1 = float(expense1_str.replace('$', ''))

expense2_str = input("Enter expense 2: ")
expense2 = float(expense2_str.replace('$', ''))

expense3_str = input("Enter expense 3: ")
expense3 = float(expense3_str.replace('$', ''))

total_expenses = expense1 + expense2 + expense3
remaining_budget = income - total_expenses

print("\nüìä BUDGET SUMMARY")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

if remaining_budget < 0:
    print("‚ö†Ô∏è Warning: You are over budget!")
else:
    print("‚úÖ You are within your budget!")
```
